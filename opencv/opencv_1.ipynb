```
import cv2 
from google.colab.patches import cv2_imshow
img=cv2.imread('/content/jurassic.jpg')
cv2_imshow(img)
cv2.waitKey(0)   # pause the execution 0 represents press any key  to resume
```

```
img.shape   # height,width,channel
```

**Resizing the image ,keeping the ratio of the image proportional to the old image width and height , so r= ratio of width of new to the old image**

```
r=100.0 / img.shape[1]    # new width / old width = ratio 
dim=(100,int(img.shape[0]*r))  # width , height of old image * ratio  # always we have to represent image as w x h 
```

```
resized=cv2.resize(img,dim,interpolation=cv2.INTER_AREA) # interpolation s the algorithm used to resize
cv2_imshow(resized)
```

```
resized.shape   # height , width , channel
```

**Flipping the image uoside down**

```
(h,w)=img.shape[:2]     
center=(w/2,h/2)  # calculating the centr by dim of the image 

m=cv2.getRotationMatrix2D(center,90,0.5)  # 180 degree , this will return matrix,and  1,0 is the scaling factor 
rotated=cv2.warpAffine(img,m,(w,h))   # actual rotation 

cv2_imshow(rotated)
```

**Cropping the image**

```
cropped=img[100:250,350:550]
cv2_imshow(cropped)
cropped.shape
```

**saving the image**

```
cv2.imwrite("cropped.jpg",cropped)
```
